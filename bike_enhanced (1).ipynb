```python
# Ensure appropriate indentations and readability
df.drop(["instant","dteday"],axis=1,inplace=True)
```

```python
# Ensure appropriate indentations and readability
# Split the data into training and testing sets
# This step ensures the model is evaluated on unseen data
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# Ensure appropriate indentations and readability
from sklearn.preprocessing import MinMaxScaler
```

```python
# Ensure appropriate indentations and readability
scaler = MinMaxScaler()
```

```python
# Ensure appropriate indentations and readability
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ["temp","atemp","hum","windspeed","casual","registered","season","yr","mnth","weekday","weathersit","cnt"]

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
# Ensure appropriate indentations and readability
df_train.head()
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
# Train the model using the training dataset
# Make sure to evaluate performance after fitting
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train[['registered']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
lr.params
```

```python
# Ensure appropriate indentations and readability
# Generate visualizations for exploratory data analysis
# Visuals help understand trends and patterns in the data
# Let's visualise the data with a scatter plot and the fitted regression line
plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], 0.024532 + 0.933473*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
# Ensure appropriate indentations and readability
print(lr.summary())
```

```python
# Ensure appropriate indentations and readability
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Ensure appropriate indentations and readability
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Dropping the variable and updating the model

```python
X = X_train.drop(['holiday','atemp',"season","hum",'temp',"mnth","windspeed","workingday"], axis=1)
```

```python
# Ensure appropriate indentations and readability
X_train_lm = sm.add_constant(X)

lr_2 = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Ensure appropriate indentations and readability
print(lr_2.summary())
```

```python
# Ensure appropriate indentations and readability
# Calculate the VIFs again for the new model

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Ensure appropriate indentations and readability
X_train_lm.head()
```

```python
# Ensure appropriate indentations and readability
y_train_cnt = lr_2.predict(X_train_lm)
```

## Step 7: Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
# Generate visualizations for exploratory data analysis
# Visuals help understand trends and patterns in the data
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

## Step 9: Model Evaluation

```python
num_vars = ["temp","atemp","hum","windspeed","casual","registered",
            "season","yr","mnth","weekday","weathersit","cnt"]

df_test[num_vars] = scaler.transform(df_test[num_vars])
df_test.describe()
```

```python
y_test = df_test.pop('cnt')
X_test = df_test[["yr","weekday","weathersit","casual","registered"]]

X_test_m = sm.add_constant(X_test)
```

```python
X_test_m.head()
```

```python
y_test_pred = lr_2.predict(X_test_m)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
